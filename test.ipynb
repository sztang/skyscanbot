```python
import datetime
base = datetime.datetime.strptime('2021-06-18','%Y-%m-%d')
datelist = [base + datetime.timedelta(days=x) for x in range(25)]
datelist
```

```python
len(datelist)
```

```python
datelist[0].strftime('%Y-%m-%d')
```

```python
def datepairs(datelist,duration): #returns start/end date pairs within a range of dates, x-days apart, in %Y-%m-%d string format
    datepairlist = []
    for d in datelist:
        d1 = d + datetime.timedelta(days=duration)
        if d1 not in datelist:
            break
        datepairlist.append((d.strftime('%Y-%m-%d'),d1.strftime('%Y-%m-%d')))
    return datepairlist
```

```python
dpairs = datepairs(datelist,10)
```

```python
count = 0
for x in range(9,15):
    datepair = datepairs(datelist,x)
    for d in datepair:
        count+=1
count
```
